# 載入套件

```
import requests
import json
import csv
```

# 抓取網址
此網址為台積電股票資訊


```
s=requests.get("https://www.twse.com.tw/exchangeReport/STOCK_DAY?response=json&date=20190720&stockNo=2330&_=1563609038052")
jd = json.loads(s.text)
jd
```

# 匯出檔案

```
outputfile = open (r'D:\Python\抓取股票.csv',"w",newline='')
outputwriter = csv.writer(outputfile)
outputwriter.writerow(jd['title'])
outputwriter.writerow(jd['fields'])
for data in (jd['data']):
    outputwriter.writerow(data)
outputfile.close()
print("檔案儲存至D:\Python\抓取股票.csv")
```
